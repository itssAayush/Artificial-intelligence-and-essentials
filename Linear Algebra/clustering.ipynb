```python
#k-means clustering example (Simple numerical)

import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Dataset

data=np.array([[2],[4],[10],[12]])

#apply k-means with k=2

kmeans=KMeans(n_clusters=2,random_state=0, n_init='auto')
kmeans.fit(data)

#cluster labels
labels=kmeans.labels_

#centroids
centroids=kmeans.cluster_centers_

print("Data points:",data.flatten())
print("Cluster labels:",labels)
print("Centroids:",centroids)

#visualisation

plt.scatter(data,[0]*len(data),c=labels,cmap='viridis',s=100)
plt.scatter(centroids,[0]*len(centroids),c='red',marker='x',s=200,label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Data')
plt.ylabel('Cluster')
plt.legend()
plt.show()
```
