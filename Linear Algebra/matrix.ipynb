```python
import numpy as np
```

```python
print("****1.Dot Product****")
x=np.array([1,2,3])
y=np.array([4,5,6])
dot_product=np.dot(x,y)
print("Dot product:",dot_product)

print("*****2.Cross Product*****")
x=np.array([1,2,3])
y=np.array([4,5,6])
cross_product=np.cross(x,y)
print("Cross product:",cross_product)

print("*****3.Matrix Multiplication*****")
x=np.array([[1,2],[3,4]])
y=np.array([[5,6],[7,8]])
matrix_product=np.matmul(x,y)
print("Matrix product:\n",matrix_product)

print("*****4.Inverse of a matrix*****")
x=np.array([[1,2],[3,4]])
inverse_matrix=np.linalg.inv(x)
print("Inverse of matrix:\n",inverse_matrix)

print("****Tranpose of a matrix****")
x=np.array([[1,2],[3,4]])
transpose_matrix=np.transpose(x)
print("Transpose of matrix:\n",transpose_matrix)

print("*****5.Eigenvalues and Eigenvectors*****")
x=np.array([[1,2],[3,4]])
eigenvalues,eigenvectors=np.linalg.eig(x)
print("Eigenvalues:",eigenvalues)
print("Eigenvectors:\n",eigenvectors)

print("*****6.Determinant of a matrix*****")
x=np.array([[1,2],[3,4]])
determinant=np.linalg.det(x)
print("Determinant of matrix:",determinant)

print("*****7.Matrix addition and subtraction*****")
x=np.array([[1,2],[3,4]])
y=np.array([[5,6],[7,8]])
addition=x+y
subtraction=x-y
print("Matrix addition:\n",addition)
print("Matrix subtraction:\n",subtraction)

print("\n -----Solve Linear Equations-----")
# 2x + y = 5
# x + y = 3
A = np.array([[2,1],
              [1,1]])
B = np.array([5, 3])

solution = np.linalg.solve(x,y)
print(solution)
```
