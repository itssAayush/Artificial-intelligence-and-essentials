```python
#supervised machine learning example
#linear regression using scikit learn

#1.import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#2.create sample dataset(house size vs price)
#x=feature(house size in sq ft)
#y=target(price)
x=np.array([500,800,1000,1200,1500,1800,2000]).reshape(-1,1)
y=np.array([150,200,250,300,350,400,450])

#3.split into training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#4.create linear regression model
model=LinearRegression()

#5.train model on training data
model.fit(x_train,y_train)

#6.make predictions
y_pred=model.predict(x_test)

#7.evaluate model
mse=mean_squared_error(y_test,y_pred)
print("model coefficinet(slope): ",model.coef_[0])
print("model intercept(bias): ",model.intercept_)
print("mean squared error: ",mse)

#8.predict new value
new_house_size=np.array([[1700]])
predicted_price=model.predict(new_house_size)
print("predicted price for 1700 sq ft house:",predicted_price[0])

#9.plot results
plt.scatter(x,y,color='blue',label='Actual data')
plt.plot(x,model.predict(x),color='red',label='Regression line')
plt.xlabel("House size(sq ft)")
plt.ylabel("Price")
plt.title("House price prediction")
plt.legend()
plt.show()
```
